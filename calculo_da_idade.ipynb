```python
#usando função input#
# Solicita a idade e o ano de nascimento e calcula o ano atual com base na idade e no ano de nascimento
idade = int(input("Digite a sua idade atual : "))
ano_nascimento = int(input("Digite o ano que você nasceu: "))

# Calcula o ano atual com base na idade e no ano de nascimento
ano_atual = ano_nascimento + idade
print("O ano atual calculado é %d" % ano_atual)

# Alternativamente, se você quiser calcular a idade com base no ano atual, aqui está um exemplo:
ano_atual = int(input("Digite o ano atual: "))
idade_calculada = ano_atual - ano_nascimento
print("Sua idade calculada é %d" % idade_calculada)

```
