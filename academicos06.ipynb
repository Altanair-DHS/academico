```python
import pandas as pd
```

```python
alunosDic = {'nomes':['ricado','pedro','joão','afonso'],'nota':[4,5.5,8,9],'aprovados':['não','não','sim','sim']}
alunosdf = pd.DataFrame( alunosDic)
```

```python
print(alunosdf)
```

```python
print(alunosDic)
```

```python
alunosdf.head() #nesse comando nos mostra as primeiras linhas desse DataFrame
```

```python

alunosdf.shape # nos mostra nesse comando shape a quantidade de linhas e colunas 
```

```python
alunosdf.describe()# nesse comando informa a quantidade de dados, desvio padrão,nos mostra a media desses dados ,valor minimo , percentil minino,maximo.
```

```python
#Filtrando linhas e colunas no Pandas
#Objetivos: aprender como realizar buscas e atribuições específicas com linhas e
#com colunas nos dataframes criados com o pacote Pandas. Essas técnicas são
#também conhecidas como slicing (do inglês “fatiamento”).
```

```python
alunosdf.loc[[]] #nesse comando eu encontro o nome das colunas#
```

```python
alunosdf.loc[0:2] #posso localiza todas as linha e colunas usando a função "loc"
```

```python
alunosdf.loc[ alunosdf['nomes'] == 'afonso']# nesse formato voce traz apenas a linha que tem o nome - nota - aprovado
```

```python
alunosdf.loc[ alunosdf['aprovados'] == 'sim']
```
