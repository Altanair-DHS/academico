```python
kmh = [78,50 ,60, 30,20,80,90,100,256 ] # lista kilomtros por horas
```

```python
kph = [40, 60, 80, 100]  # velocidades em km/h

mph = []  # lista vazia

for i in kph:
    mph.append(i / 1.20)

print(mph)
```
