```python
import time

contador = 0

while contador < 10:
    print('ainda não deu! mas dará certo')

    if contador == 3:
        break  # interrompe o laço quando chega em 3

    contador = contador + 1
    time.sleep(1)  # pausa de 1 segundo entre as repetições

print('agora foi de show')

```
